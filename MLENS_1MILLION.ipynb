Pearson Based Movie Similarity
--- 

Putting the above elements together I can now create a Pearson correlation-based moveie similarity function 


```python
def moviePairSimilarityPearson(movx, movy, rs,verbose=False):
    m1 = rs.ix[movx].reset_index() 
    m2 = rs.ix[movy].reset_index() 
    mmerge = pd.merge(m1,m2,on='uid').drop('uid',axis=1)
    mmerge['rating_x'] = mmerge['rating_x'].astype('float32')
    mmerge['rating_y'] = mmerge['rating_y'].astype('float32')
    pcorr = mmerge.corr(method='pearson')
    if verbose: print movx, movy, mmerge
    return pcorr.ix['rating_x']['rating_y']
```

Adjusted Cosine Based Similarity 
--- 

This is a variation based on *Serwer et al.* http://www.ra.ethz.ch/cdstore/www10/papers/pdf/p519.pdf using the adjusted cosine function. 
$$ \frac{ \sum__{u \in U} (R_{u,i} - \bar{R_u})(R_{u,j} - \bar{R_u})}{\sqrt{ \sum (R_{u,i} - \bar{R_u})^2}\sqrt{ \sum (R_{u,j} - \bar{R_u})^2} } $$


```python
def moviePairSimilarityAdjCos(movx, movy, rs,verbose=False):
    m1 = rs.ix[movx].reset_index() 
    m2 = rs.ix[movy].reset_index() 
    mm = pd.merge(m1,m2,on='uid').drop('uid',axis=1)
    num = (mm['rating_x'] - mm['avgRating_x']).dot((mm['rating_y'] - mm['avgRating_y']))
    den1 = (mm['rating_x'] - mm['avgRating_x']).dot((mm['rating_x'] - mm['avgRating_x']))
    den2 = (mm['rating_y'] - mm['avgRating_y']).dot((mm['rating_y'] - mm['avgRating_y']))

    sim = num/numpy.sqrt(den1*den2)
    if sim > 1.0: sim = 1.0 
    return sim 
```

```python
import numpy
def bestMatches(movx, movie_tls,rs,nTop=100, simFunc = moviePairSimilarityPearson):
        score = []
        for movy in movie_tls:
            #print movy
            if movx == movy: continue
            corr = simFunc(movx, movy, rs)
            if pd.isnull(corr): continue
            score += [(corr, movy.decode('unicode-escape'))] 
        score.sort()
        score.reverse()
        return score[0:nTop]
```

Similarity Matrix 
----

Now, we build the item-item similarity matrix. This can be a long job, so I would do it in chunks over multiple processors. The `iStart` and `iStop` params
let me chose the chunk I want to work on.  Note that I use a unicode-decode option so that the table can later be dumped in a JSON format. 


```python
def makeSimilarityTable(movie_tls, rs,n=1000, simFunc = moviePairSimilarityAdjCos,iStart=0, iStop=None):
	scores = {}
	if iStop == None:
		iStop = len(movie_tls)

	for movx in movie_tls[iStart:iStop]:
		score = bestMatches(movx, movie_tls,rs,nTop=n, simFunc = simFunc)
		scores[movx.decode('unicode-escape')] = score

	return scores 
```

```python
import pickle
def recommendMovies(userRatings,simTable):
    
    scores = {} 
    totalSim = {} 
    hits = {}
    for (item, rating) in zip(userRatings.title, userRatings.rating): 
        for (similarity, item2) in simTable[item]:
            if item2 in userRatings: continue 
            scores.setdefault(item2,0) 
            scores[item2] += similarity*rating 
            totalSim.setdefault(item2,0) 
            totalSim[item2] += similarity
            hits.setdefault(item2,0) 
            hits[item2] += 1
    for tit, hit in hits.items():
        if hit < 20:
            scores.pop(tit)
            hits.pop(tit)
            totalSim.pop(tit)
    
    rankings = [(score/totalSim[item], item) for item, score in scores.items()]
    rankings.sort()
    rankings.reverse()
    return rankings, hits 
```
